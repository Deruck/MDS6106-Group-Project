```python
import os
os.environ["KMP_DUPLICATE_LIB_OK"]="TRUE"
%matplotlib inline

# 读取数据
from tools.datasets import DataGenerator
data_generator=DataGenerator(p=3, sep=10,seed=4602)
data,label=data_generator.get_data()
data_generator.plot()
```

```python
# 设定参数
general_param={
    'data':data,
    'lambda_': 0.5
}

model_param={
    'model':'WeightedL1',
    'k':20,
    'v':0.5,
    'delta':0.01
}

minimizer_param={
    'minimizer': 'LBFGS',
    'tol': 10**(-5),
    's': 1,
    'm': 20,
    'gamma': 0.1,
    'sigma': 0.5,
    'max_line_search_iter': 10
}


# 加载模型
from tools.full_model import MyCluster
my_cluster=MyCluster(**general_param,**model_param,**minimizer_param)

# 优化
my_cluster.optimize()

# 聚类
my_cluster.solution_to_cluster(epsilon=0.01)

# 获取结果看板（beta控制vmeasure的加权）
my_cluster.get_result_panel(label,beta=1)

# 聚类结果可视化（仅生成二维数据可用）
my_cluster.plot2d(data_generator)
```
