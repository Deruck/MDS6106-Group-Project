### 1.1 生成数据

```
class DataGenerator: 用于生成数据的类
    __init__:
        p[int]: 类个数
        field[list(2 x 2)]: 两个维度的定义域，默认1-10，1-10
        np_list [list < int >]: 每个类点个数, 默认每类20个点
        refernce_point_list [list < list < float >>]: 每个类的中心，默认在定义域上从均匀分布中随机生成
        sd_list [list < float >]: 每个类的方差，默认为每一类离最近点距离/分离度(sep)
        sep[float]: 类间数据分离度
        seed[int]: 随机数种子，默认随机生成

    Attributes:
        seed[int]: 当前随机数种子

        _data[scipy.sparse.csc.csc_matrix (2 x n)]: 
        _label[np.array (n,)]:

    Method:
        get_data(): 获取数据
        plot(): 绘制数据图像
        write(id[str]): 将数据写入文件

        _generate_data(): 生成数据
```

````python
from tools.datasets import DataGenerator
````

使用方法：
- 初始化`DataGenerator`类并传入参数，生成数据
- 通过`DataGenerator.get_data()`方法获取生成数据

````python
data_generator=DataGenerator(p=5,sep=4)
data,label=data_generator.get_data()
print("dataset\tdata_shape\tlabel_shape\tdata_type\tlabel_type\t")
print(f"self-generated\t{data.shape}\t{label.shape}\t{type(data)}\t{type(label)}")
````

可以使用`DataGenerator.plot()`方法查看生成数据图像。

````python
data_generator.plot()
````

`p`参数控制类的个数。

````python
for p in range(2,5):
    print(f'p={p}')
    data_generator=DataGenerator(p=p,sep=4)
    data_generator.plot()
````

没有人为指定方差时，`sep`参数可以控制类间的分离程度。

````python
seed=1
for sep in [1,3,5,8]:
    print(f'sep={sep}')
    data_generator=DataGenerator(p=5,sep=sep,seed=seed)
    data_generator.plot()
````

如果随机得到一组比较满意的数据，可以通过`DataGenerator.seed`属性获得当前随机数种子，方便复现。

````python
data_generator=DataGenerator(p=5,sep=4)
seed=data_generator.seed
print(f'seed={seed}')
data_generator.plot()
````

````python
data_generator=DataGenerator(p=5,sep=4,seed=seed)
data_generator.plot()
````

也可以人为指定每一类的样本量、方差以及中心。

````python
params={
    'p':5,
    'sep':4,
    'np_list':[5,6,7,8,9],
    'sd_list':[0.3]*5,
    'refernce_point_list':[[1,0],[3,0],[4,2],[5,3],[7,4]],
    'seed':10
}

data_generator=DataGenerator(**params)
data_generator.plot()
````

可以使用`DataGenerator.write(id[str])`方法将生成数据写入文件，可用下面的`load_data`函数读取。
- 存入`data`文件夹中.
- 命名格式：`self_generated_{id}_{data|label}`

````python
data_generator.write('1')
````

## 2 函数
`tools\functions.py`

- 函数可以通过`FUNCTION.obj(x)`,`FUNCTION.grad(x)`,`FUNCTION.hessian(x)`获取函数的值、梯度与hessian。
- 优化目标函数可以通过定义基本部件，然后线性组合。支持运算符：
  - 左乘常数
  - 右除以常数
  - 两函数相加
- 编写函数方法类比`tools\functions.py`中`SSE`函数的编写。

````python
from tools.functions import SSE
from tools.functions import Huber
from tools.datasets import coo_to_tensor
````

````python
data_generator=DataGenerator(p=5,sep=4,seed=9408)
data,label=data_generator.get_data()
data_shape=data.shape
# data=data.reshape(-1,1)

# from scipy.sparse import csc_matrix
# import numpy as np
# x=csc_matrix(np.ones(data.shape))

from scipy.sparse import coo_matrix
import numpy as np
x = coo_matrix(np.ones(data.shape))
x = coo_to_tensor(x)
````

初始化`SSE`函数，传入数据。

$$\begin{aligned}
SSE(x|\alpha)&=(x-\alpha)^T(x-\alpha)\\
\nabla SSE(x|\alpha)&=2(x-\alpha)\\
\nabla^2 SSE(x|\alpha)&=2I_{dn}
\end{aligned}$$

````python
sse=SSE(data)

print(sse.obj(x))
print(sse.grad(x).shape)
print(sse.hessian(x).shape)
````

````python
x = x.to_dense()
````

````python
np.identity(2)
````

可通过
$$f_{cluster}(x)={1\over2}SSE(x)+\lambda\phi_{hub}(x)$$
来获得最终优化目标函数

````python
huber=Huber(delta=1, data_shape=data_shape)
lbd=0.1
f_cluster=sse/2+lbd*huber
f_cluster.obj(x)
````

## 3 优化器
`tools\optimization.py`

````python
from tools.datasets import DataGenerator

data_generator = DataGenerator(p=5, sep=4, seed=9408)
data, label = data_generator.get_data()
data = data.to_dense()
data = data.reshape(-1,1)
````

优化器父类
```
class Minimizer:
    __init__:
        function[function]: 优化目标函数
        init[np.array(nd x 1)]: 初始点
        tol[float]: 终止条件
        max_iter[int]: 最多迭代次数

    Attributes
        _function:
        _init:
        _tol:
        _solution[array(init.shape)]: 优化解，和起始点同shape
        _duration[float]: 计算时间
        _iter[int]: 迭代次数
        _path[list<np.array>]: 优化路径

    Method
        get_result: 获取结果

        _apply: 实施优化
        _line_search: 步长策略
```
其他优化器继承`Minimizer`类，编写方式参照示例`tools.optimization.ExampleMinimizer`。

**示例：**<br>
`tools.optimization.ExampleMinimizer`
- `function`, `init`, `tol`, `max_iter`为通用参数
- `param1`,`param2`为该优化器特有参数

````python
from tools.optimization import ExampleMinimizer, AGM
from scipy.sparse import csc_matrix
from tools.functions import SSE
import numpy as np

%matplotlib

f=SSE(data)/2
init=csc_matrix(np.ones(data.shape))
tol=10**(-3)
max_iter=1000
param={
    'param1': 1,
    'param2': 2    
}
# example_minimizer=ExampleMinimizer(function=f,init=init,tol=tol,max_iter=max_iter,**param)
example_minimizer=AGM(function=f,init=init,tol=tol,max_iter=max_iter,**param)
````

````python
result=example_minimizer.get_result()
result.keys()
````
